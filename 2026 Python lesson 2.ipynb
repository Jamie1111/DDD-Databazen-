### Creating statistic of your data

Creating statisitics of your data is very usefull if you want to get to knwo your data. There are a lot of statistics you can calaculate with your data like the average, median, mode and standard deviation. All these statistics already have functions in Pandas which means you do not have to calculate them yourslef :).

#### Mean
The average of all the values in the column. Useful for example if you want to know what the average amount of sales is globally. This can be done by using the `mean()` function on the **Global_Sales** column.

```python
df['Global_Sales'].mean()
```

##### Min and max
Getting the lowest value in a column is useful to know because this allows you to get an idea of the range of values together with the maximum value. Use the `min()` adn `max()` functions to get these values.

```python
print(df['Global_Sales'].min())
print(df['Global_Sales'].max())
```

#### Median
This is the value precisely in at 50% after sorting all value in order from small to large. Meaning about 50% of the data is above and below this value. You can use this to get a better understanding of the average value. Because average might be skewed by extremely high or low values. You can calculate the median by using the `median()` function.

```python
df['Global_Sales'].median()
```

#### Std
The deviation of each value in the column from the mean, taking the square root of those values and save the results in a list. Calculate the mean for that list of deviations and the take the root. Good to get an idea of the dispersion of the values in a column, you can do this by using the `std()` function.

```python
df['Critic_Score'].std()
```

#### Mode
The mode function gives you the most common value in the column. So for example if we would like to see what the most given score is by critics we can do the `mode()` function on the column **Critic_Score**.

```python
df['Critic_Score'].mode()
```

#### Describe

This function gives you all the statistics in one overview for the **numeric** columns.

```python
df.describe()
```

#### Unique

This function gives you all the unique value in a column.

```python
df['Platform'].unique()
```

#### Nunique

### Boxplot

A **boxplot** is a compact way to visualize the distribution of numerical data and quickly spot patterns, spread, and outliers. It summarizes a dataset using five key values: the **minimum**, **first quartile (Q1)**, **median**, **third quartile (Q3)**, and **maximum**.

The **box** shows the interquartile range (IQR), which contains the middle 50% of the data, while the **line inside the box** marks the median. The **whiskers** extend to the smallest and largest values within a chosen range, and any points beyond them are plotted individually as **outliers**.

In Python, boxplots are commonly created with **Matplotlib** and are especially useful for comparing distributions across multiple groups.

```python
plt.boxplot(df['User_Score'].dropna())
plt.show()
```

### 3 Filtering & Selection
Instead of showing a random sample, you might want to have a work with a specific row, column, or a selected group. This works in a similar fashion as selecting keys in a dictionary. 

One of the most important skills to have is to be able to filter your dataframe based on a statement. E.g., let's gather all the Xbox 360 games.

Use square brackets to select a column - `df["Platform"]` and `loc[]` to filter the dataframe based on certain conditions.

#### Examples:

```python
df.loc[df["Platform"] == "X360"]
```

Here are some more advanced filtering statements

```python
# Find games released on the PS4 and sold more than 1 million copies globally
df.loc[(df["Platform"] == "PS4") & (df["Global_Sales"] > 1)]
```

```python
# Identify the name and critic score of action games released in 2008 that had a critic score of more than 90
df.loc[(df["Genre"] == "Action") & (df["Year_of_Release"] ==2008) & (df["Critic_Score"] > 90), ["Name", "Critic_Score"]]
```

### Select
Another way to filter your dataframe is by using the `iloc[]` function. This [function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html) is purely used for integer-location based indexing for selecting by position.

In other words, rows or columns are selected based on their location in the dataframe. Before you can use this, you need to understand how selection selecting works first.

```python
# First 4 rows and all the columns
df.iloc[0:4]
df.iloc[0:4,:]
```

```python
# All the rows and first 4 columnsS
df.iloc[:,0:4]
```

```python
# First 4 rows and first 4 columns
df.iloc[0:4,0:4]
```

```python
# First 4 rows and third + fourth column
df.iloc[0:4,2:4]
```

```python
# Second + third row and third + fourth column
df.iloc[1:3,2:4]
```
